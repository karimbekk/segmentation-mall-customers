```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import warnings
warnings.filterwarnings('ignore')

# Configuration pour les graphiques
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("✅ Toutes les librairies importées avec succès")
```

```python
# Charger le dataset
df = pd.read_csv('Mall-Customers.csv')

# Afficher les premières lignes
print("Premières lignes du dataset :")
print(df.head(10))

# Informations générales
print("\n" + "="*60)
print("Informations du dataset :")
print(df.info())

print("\n" + "="*60)
print("Statistiques descriptives :")
print(df.describe())

print("\n" + "="*60)
print(f"Dimensions : {df.shape}")
print(f"Valeurs manquantes :\n{df.isnull().sum()}")
```

```python
# Distribution de l'âge
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Âge')
plt.ylabel('Nombre de clients')
plt.title('Distribution de l\'âge')

# Distribution du revenu
plt.subplot(1, 3, 2)
plt.hist(df['Annual Income (k$)'], bins=20, color='lightcoral', edgecolor='black')
plt.xlabel('Revenu annuel (k$)')
plt.ylabel('Nombre de clients')
plt.title('Distribution du revenu annuel')

# Distribution du score de dépense
plt.subplot(1, 3, 3)
plt.hist(df['Spending Score (1-100)'], bins=20, color='lightgreen', edgecolor='black')
plt.xlabel('Score de dépense')
plt.ylabel('Nombre de clients')
plt.title('Distribution du score de dépense')

plt.tight_layout()
plt.show()

print("Statistiques détaillées des 3 features clés :")
print(df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].describe())
```

```python
# Sélectionner les features pour le clustering
X = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].copy()

print("Features sélectionnées :")
print(X.head(10))

print("\n" + "="*60)
print("Valeurs AVANT scaling :")
print(f"Age : min={X['Age'].min()}, max={X['Age'].max()}")
print(f"Annual Income : min={X['Annual Income (k$)'].min()}, max={X['Annual Income (k$)'].max()}")
print(f"Spending Score : min={X['Spending Score (1-100)'].min()}, max={X['Spending Score (1-100)'].max()}")

print("\n💡 RAISON DU SCALING :")
print("K-means utilise la distance euclidienne.")
print("Sans scaling, le revenu (0-200) dominerait l'âge (18-70).")
print("On doit normaliser pour que chaque variable ait le même poids.")

# Standardiser les données (scaling)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("\n" + "="*60)
print("Valeurs APRÈS scaling (moyenne ≈ 0, écart-type ≈ 1) :")
print(f"Moyenne : {X_scaled.mean(axis=0)}")
print(f"Écart-type : {X_scaled.std(axis=0)}")
print(f"\nShape de X_scaled : {X_scaled.shape}")

# X_scaled est un numpy array, on le remet en DataFrame pour bien voir
X_scaled_df = pd.DataFrame(
    X_scaled, 
    columns=['Age_scaled', 'Income_scaled', 'Score_scaled']
)

# Afficher les 10 premières lignes après scaling
print(X_scaled_df.head(10))

# Vérifier les stats après scaling
print("\nStatistiques après scaling :")
print(X_scaled_df.describe())

```

```python
# Calculer l'inertie (WCSS) pour différentes valeurs de k
inertias = []
silhouette_scores = []
K_range = range(2, 11)

print("Calcul en cours pour k = 2 à 10...")
for k in K_range:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

print("✅ Calcul terminé")

# Tracer les graphiques
plt.figure(figsize=(14, 5))

# Graphique 1 : Courbe du coude
plt.subplot(1, 2, 1)
plt.plot(K_range, inertias, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Nombre de clusters (k)', fontsize=12)
plt.ylabel('Inertia (WCSS)', fontsize=12)
plt.title('Méthode du Coude - Inertia', fontsize=13, fontweight='bold')
plt.grid(True, alpha=0.3)

# Graphique 2 : Silhouette score
plt.subplot(1, 2, 2)
plt.plot(K_range, silhouette_scores, 'ro-', linewidth=2, markersize=8)
plt.xlabel('Nombre de clusters (k)', fontsize=12)
plt.ylabel('Silhouette Score', fontsize=12)
plt.title('Silhouette Score par k', fontsize=13, fontweight='bold')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Afficher les valeurs
print("\n" + "="*60)
print("📊 RÉSULTATS DÉTAILLÉS :")
print("="*60)
for k, inertia, silhouette in zip(K_range, inertias, silhouette_scores):
    print(f"k = {k:2d} → Inertia = {inertia:8.2f} | Silhouette Score = {silhouette:.4f}")

print("\n💡 INTERPRÉTATION :")
print("- Inertia : plus bas = meilleur, mais chaque k supplémentaire aide moins")
print("- Le 'coude' = le point où la courbe s'aplatit (k optimal)")
print("- Silhouette Score : plus proche de 1 = meilleur, mesure la qualité des clusters")
```

```python
# D'après la courbe du coude et silhouette score, choisir k
optimal_k = 6

print(f"✅ K optimal choisi : {optimal_k}")
print(f"   Silhouette Score pour k={optimal_k} : {silhouette_scores[optimal_k-2]:.4f}")
print(f"   Inertia pour k={optimal_k} : {inertias[optimal_k-2]:.2f}")
```

```python
# Entraîner le modèle final avec k optimal
kmeans_final = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42, n_init=10)
clusters = kmeans_final.fit_predict(X_scaled)

# Ajouter les labels de cluster au dataframe original
df['Cluster'] = clusters

print(f"✅ K-means appliqué avec k={optimal_k}")
print(f"\nLabels des clusters : {np.unique(clusters)}")

print("\n" + "="*60)
print("Nombre de clients par cluster :")
print(df['Cluster'].value_counts().sort_index())

print("\nPourcentage par cluster :")
for cluster in sorted(df['Cluster'].unique()):
    count = len(df[df['Cluster'] == cluster])
    pct = count / len(df) * 100
    print(f"Cluster {cluster} : {count} clients ({pct:.1f}%)")
```

```python
# Scatter plot : Revenu vs Spending Score (coloré par cluster)
plt.figure(figsize=(10, 7))

# Plot les points par cluster
colors = plt.cm.Set3(np.linspace(0, 1, optimal_k))
for cluster in range(optimal_k):
    cluster_points = df[df['Cluster'] == cluster]
    plt.scatter(cluster_points['Annual Income (k$)'], 
                cluster_points['Spending Score (1-100)'],
                label=f'Cluster {cluster}',
                s=100,
                alpha=0.6,
                edgecolors='black',
                linewidth=0.5,
                c=[colors[cluster]])

# Plot les centroïdes (conversion de l'espace scaling → espace original)
centroids = scaler.inverse_transform(kmeans_final.cluster_centers_)
plt.scatter(centroids[:, 1],  # Annual Income (colonne 1)
            centroids[:, 2],  # Spending Score (colonne 2)
            c='red',
            marker='X',
            s=500,
            edgecolors='black',
            linewidth=2,
            label='Centroïdes',
            zorder=5)

plt.xlabel('Annual Income (k$)', fontsize=12)
plt.ylabel('Spending Score (1-100)', fontsize=12)
plt.title(f'Customer Segmentation avec K-means (k={optimal_k})', fontsize=13, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("✅ Visualisation 2D créée (Revenu vs Spending Score)")
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

colors = plt.cm.Set3(np.linspace(0, 1, optimal_k))
for cluster in range(optimal_k):
    cluster_points = df[df['Cluster'] == cluster]
    ax.scatter(cluster_points['Age'],
               cluster_points['Annual Income (k$)'],
               cluster_points['Spending Score (1-100)'],
               label=f'Cluster {cluster}',
               s=50,
               alpha=0.6,
               c=[colors[cluster]])

ax.set_xlabel('Age', fontsize=11)
ax.set_ylabel('Annual Income (k$)', fontsize=11)
ax.set_zlabel('Spending Score (1-100)', fontsize=11)
ax.set_title(f'Customer Segmentation en 3D (k={optimal_k})', fontsize=13, fontweight='bold')
ax.legend(fontsize=10)
plt.show()

print("✅ Visualisation 3D créée (Age vs Revenu vs Spending Score)")
```

```python
# Calculer la moyenne de chaque feature par cluster
cluster_summary = df.groupby('Cluster')[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].mean()

print("="*70)
print("📊 RÉSUMÉ DES CLUSTERS (moyennes)")
print("="*70)
print(cluster_summary.round(2))

print("\n" + "="*70)
print("📊 ANALYSE DÉTAILLÉE DE CHAQUE CLUSTER")
print("="*70)

for cluster in range(optimal_k):
    cluster_data = df[df['Cluster'] == cluster]
    n_clients = len(cluster_data)
    pct = n_clients / len(df) * 100
    
    print(f"\n🎯 CLUSTER {cluster} ({n_clients} clients, {pct:.1f}% du total)")
    print("-" * 70)
    print(f"   Âge moyen : {cluster_data['Age'].mean():.1f} ans (min: {cluster_data['Age'].min()}, max: {cluster_data['Age'].max()})")
    print(f"   Revenu moyen : ${cluster_data['Annual Income (k$)'].mean():.1f}k (min: ${cluster_data['Annual Income (k$)'].min()}, max: ${cluster_data['Annual Income (k$)'].max()})")
    print(f"   Score de dépense moyen : {cluster_data['Spending Score (1-100)'].mean():.1f}/100 (min: {cluster_data['Spending Score (1-100)'].min()}, max: {cluster_data['Spending Score (1-100)'].max()})")
    
    gender_counts = cluster_data['Gender'].value_counts()
    print(f"   Répartition sexe : {dict(gender_counts)}")
```

```python
# 9. PRÉDICTION NOUVEAU CLIENT
print("\n" + "="*70)
print("🔮 PRÉDICTION NOUVEAU CLIENT")
print("="*70)

# Exemple client
new_client = np.array([[30, 85, 20]])  # Âge 30, 85k$, dépense 20
new_scaled = scaler.transform(new_client)
predicted_cluster = kmeans_final.predict(new_scaled)[0]

print(f"👤 NOUVEAU CLIENT : 30 ans, 85k$, dépense 20")
print(f"🎯 CLUSTER PRÉDIT : {predicted_cluster}")
print(f"📊 Profil : {cluster_interpretations[predicted_cluster]['nom']}")

# Distance au centroïde
distances = np.linalg.norm(X_scaled - kmeans_final.cluster_centers_[predicted_cluster], axis=1)
closest_client_idx = np.argmin(distances)
print(f"Client le plus proche : {df.iloc[closest_client_idx][['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].values}")

print("\n✅ ANALYSIS COMPLÈTE TERMINÉE !")
print("\n💾 Sauvegarde : df.to_csv('segmented_customers.csv', index=False)")
df.to_csv('segmented_customers.csv', index=False)
```
